```python
#check on December-8, 2025 by Hengcong Liu

import pandas as pd
from pathlib import Path

def combine_risk_factor(bios, period, check_na=True):
    # === 1. 定义输入输出路径 ===
    climate_file = Path(f'input/bioc_county_yearly_mean_{period}.csv')
    elevation_file = Path('input/China_county_level_mean_elevation.csv')
    landtype_file = Path('input/China_county_level_land_cover_proportions.csv')
    
    output_dir = Path('output/1_risk_factor')
    output_file = output_dir / 'risk_factor.xlsx'
    output_dir.mkdir(parents=True, exist_ok=True)  # 创建文件夹

    # === 2. 读取数据 ===
    climate = pd.read_csv(climate_file)[bios]
    elevation = pd.read_csv(elevation_file)
    landtype = pd.read_csv(landtype_file)

    # === 3. 合并数据 ===
    combine = (
        climate
        .merge(elevation, on='gb', how='left')
        .merge(landtype, on='gb', how='left')
    )

    # === 4. 检查缺失值 ===
    if check_na:
        if combine.isnull().values.any():
            print("DataFrame 中有空值")

    # === 5. 列重命名 ===
    replacements = {
        'proportion_10': 'Tree cover',
        'proportion_20': 'Shrubland',
        'proportion_30': 'Grassland',
        'proportion_40': 'Cropland',
        'proportion_50': 'Built-up',
        'proportion_60': 'Bare/sparse vegetation',
        'proportion_70': 'Snow and Ice',
        'proportion_80': 'Permanent water bodies',
        'proportion_90': 'Herbaceous wetland',
        'proportion_95': 'Mongroves',
        'proportion_100': 'Moss and lichen',
        'average_elevation': 'Elevation'
    }
    combine = combine.rename(columns=replacements)

    # === 6. 保存结果 ===
    combine.to_excel(output_file, index=False)
    print(f"风险因子数据已保存到: {output_file}")

# === 调用示例 ===
combine_risk_factor(
    bios=['gb', 'Bio3', 'Bio4', 'Bio7', 'Bio11', 'Bio12', 'Bio15', 'Bio16','Bio18'],
    period='2000_2024'
)
```
