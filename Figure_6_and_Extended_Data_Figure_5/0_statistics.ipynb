```python
import pandas as pd
from pathlib import Path

# === 1. 定义输入输出路径 ===
input_file = Path('../1_clean/data_merge_all_20260110.xlsx')
output_dir = Path('output/0_statistics')
output_file = output_dir / 'virus_county.xlsx'

# === 2. 新建输出文件夹（如果不存在） ===
output_dir.mkdir(parents=True, exist_ok=True)

df = pd.read_excel(input_file)
df = df[~(df['host_species'].isna() & df['standard virus name'].isna())].reset_index(drop=True)
print(df.shape[0])
result_all = (
    df.groupby('standard virus name', as_index=False)['gb']
    .nunique()
    .rename(columns={'standard virus name': 'virus', 'gb': 'unique_county_count_all'})
)

result_all.to_excel(output_file, index=False)

print(f"统计完成，结果已保存到: {output_file}")
```
